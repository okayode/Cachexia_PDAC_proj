```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_row', None)

# load data
# X_train = pd.read_csv('data/norm_X_train_NCa_Ca.csv')
# y_train = pd.read_csv('data/y_train_NCa_Ca.csv')

data = pd.read_csv('data/Xy_PCa_Ca_md.csv')

print(data.shape)
```

```python
# MI
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression

# ---- Data ----
X = data.drop(columns=['CACHEXSTAGE0VIG'])
y = data['CACHEXSTAGE0VIG']

# Detect task type
is_classification = y.nunique() < 20 and not np.issubdtype(y.dtype, np.floating)

# ---- Mutual Information ----
X_filled = X.fillna(0)

if is_classification:
    mi = mutual_info_classif(
        X_filled,
        y,
        discrete_features='auto',
        random_state=42
    )
else:
    mi = mutual_info_regression(
        X_filled,
        y,
        random_state=42
    )

# Sort MI scores
mi_scores = (
    pd.Series(mi, index=X.columns, name="MI")
    .sort_values(ascending=False)
)

print("\n=== Mutual Information Feature Ranking ===")
print(mi_scores.round(10))
```

```python
import matplotlib.pyplot as plt

# ---- Plot MI ----
plt.figure(figsize=(5, max(4, 0.3 * len(mi_scores))))
mi_scores.sort_values().plot(kind='barh')

plt.xlabel("Mutual Information")
plt.title("Mutual Information PCa v Ca Predictor")
plt.tight_layout()
#plt.show()

savefig('./figs/PCa_Ca_MI')
```
